# Seaborn Tutorial

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
plt.figure(figsize =(10,10))
sns.relplot(x="total_bill", y = "tip", data=tips, hue="sex")
plt.show()
```

```python
sns.relplot(x="total_bill", y = "tip", data=tips, size="tip")
plt.title("Tip Size")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
sns.relplot(x="total_bill", y = "tip", data=tips, style="size")
plt.title("Tip Size")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
sns.relplot(x="total_bill", y = "tip", data=tips, hue="size",palette=plt.cm.rainbow,size = "size", sizes = (15,200))
plt.title("Tip Size")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri);
```

```python
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri,ci=None,);
```

```python
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri,ci="sd",);
```

```python
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri, style = "event",hue = "region",markers=True,dashes=True);
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker",col="time", data=tips);
```
